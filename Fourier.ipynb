```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
%matplotlib nbagg
```

```python
X = np.linspace(0, 10*np.pi, 200)
Y = np.sin(X)
fig, ax = plt.subplots(1,1)
ax.set_xlim([0, 10*np.pi])
ax.set_ylim([-1.1, 1.1])

def sine(i):
    ax.clear()
    sinegraph, = ax.plot(X, Y)
    ax.arrow(X[i], 0, 0, Y[i], head_width=0.05, head_length=0.005, fc='k', ec='k')

    
anim = animation.FuncAnimation(fig, sine, frames=len(X), interval=50)
plt.show()
```

```python
num_of_sines = 2
scaling_coeff = 4/np.pi
X = np.linspace(0, 10*np.pi,1000)
xx = 2*np.pi*1*1j*X
Y = scaling_coeff*e**xx
xx1 = 2*np.pi*3*1j*X
Y1 = (scaling_coeff*(e**xx1))/3
xx2 = 2*np.pi*5*1j*X
Y2 = (scaling_coeff*(e**xx1))/5
xx3 = 2*np.pi*7*1j*X
Y3 = (scaling_coeff*(e**xx1))/7
fig,ax = plt.subplots(1,1)
ax.set_xlim([0, 5*np.pi])
ax.set_ylim([-10, 10])


X_real = Y.imag + Y1.imag + Y2.imag + Y3.imag
ax.plot(X, X_real)
```

```python
num_of_sines = 100
scaling_coeff = 4/np.pi
X = np.linspace(0, 10*np.pi,1000)
Y = np.array(1000)
for i in range(1, num_of_sines):
    xx = -2*np.pi*(2*i-1)*1j*X
    data = (scaling_coeff/(2*i-1))*e**xx
    Y = Y + data
fig,ax = plt.subplots(1,1)
ax.set_xlim([0, 5*np.pi])

ax.plot(X, Y.imag)
```
